```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df= pd.read_csv("Iris.csv")
```

```python
df[:5]
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("Species", axis = 1)
y = df["Species"]
```

```python
X.head(5)
```

```python
y.head(5)
```

```python
X_train,X_test, y_train, y_test =train_test_split(X, y, test_size=0.6)
```

```python
X_train
```

```python
X_test
```

```python
mdl = KNeighborsClassifier(n_neighbors=20)
mdl.fit(X_train, y_train)
```

```python
y_pred = mdl.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
knn_sonuc = accuracy_score(y_test, y_pred)
```

```python
knn_sonuc
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mdl_ = MLPClassifier(hidden_layer_sizes=(150)).fit(X_train, y_train)
```

```python
y_pred = mdl_.predict(X_test)
y_pred
```

```python
mlp_sonuc = accuracy_score(y_test, y_pred)
mlp_sonuc
```
